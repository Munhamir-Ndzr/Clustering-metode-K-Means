```python
# K-Means Clustering on Customer Data

import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
# 1. Load dataset
df = pd.read_csv("Customers.csv")
df.head()
```

```python
# 2. Pilih fitur numerik untuk clustering
X = df[['Annual Income ($)', 'Spending Score (1-100)']]
```

```python
# 3. Normalisasi data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

kmeans = KMeans(n_clusters=3, random_state=42)
df['Cluster'] = kmeans.fit_predict(X_scaled)
```

```python
# 5. Lihat hasil clustering
df[['CustomerID', 'Annual Income ($)', 'Spending Score (1-100)', 'Cluster']].head()
```

```python
# 6. Visualisasi hasil clustering
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x='Annual Income ($)', y='Spending Score (1-100)', hue='Cluster', palette='Set2', s=100)
plt.title('Hasil K-Means Clustering')
plt.xlabel('Annual Income ($)')
plt.ylabel('Spending Score (1-100)')
plt.grid(True)
plt.show()
```
